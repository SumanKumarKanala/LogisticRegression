# Logistic Regression

```python
#importing pandas library
import pandas as pd
```

```python
#loading the data into pandas Data Frame
dset=pd.read_csv('bank.csv')
```

```python
#displaying first 5 values of dataset
dset.head()
```

From the problem statement our output varible/feature is y, and all other features are our inputs

```python
#getting to know the shape of our data
dset.shape
```

```python
#getting an overview of the data
dset.info()
```

```python
#segmenting the dataset as per i/p and o/p features
x=dset.drop('y',axis=1)
x
```

As we see above x contains categorical values for several features we need to encode them using encoding methods.

we here go with label encoding because there are more than 2 categories for various features.

```python
#importing LabelEncoder from sklearn library
from sklearn.preprocessing import LabelEncoder
```

```python
#transforming the features
x['Job']=LabelEncoder().fit_transform(x.job)
x['Marital']=LabelEncoder().fit_transform(x.marital)
x['Education']=LabelEncoder().fit_transform(x.education)
x['Default']=LabelEncoder().fit_transform(x.default)
x['Housing']=LabelEncoder().fit_transform(x.housing)
x['Loan']=LabelEncoder().fit_transform(x.loan)
x['Contact']=LabelEncoder().fit_transform(x.contact)
x['Month']=LabelEncoder().fit_transform(x.month)
x['Poutcome']=LabelEncoder().fit_transform(x.poutcome)
```

```python
#now let see the x by dropping the duplicate features
x.columns
x=x.drop(['job','marital','education','default','housing','loan','contact','month','poutcome'],axis=1)
x
```

```python
xf=x.copy()
for column in xf.columns:
    xf[column]=(xf[column]-xf[column].min())/(xf[column].max()-xf[column].min())
xf
```

```python
# y or o/p feature
op=dset.iloc[:,-1]
op
```

```python
#now lets transform the y target feature
op=pd.get_dummies(op)
op
```

```python
#deleting any one column as both are inverse 
op=op.drop('no',axis=1)
op
```

```python
#renaming the column
op=op.rename({'y_yes':'y'},axis=1)
op=op.iloc[:,-1]
op
```

```python
#importing LogisticRegression model from sklearn library
from sklearn.linear_model import LogisticRegression
```

```python
#creating a instance for the model
model=LogisticRegression(solver='lbfgs', max_iter=1000)
model.fit(xf,op)
```

```python
#probability prediction on dataset
op_prob=model.predict_proba(xf)
op_prob
```

```python
#predicting the class labels
op_pred=model.predict(xf)
op_pred
```

```python
#importing model evaluaton libraries
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(op,op_pred)
cm
```

```python
#evaluation of model using confusion matrix
accuracy=((39245+1027)/(39245+677+4262+1027))*100
accuracy
```

```python
#evaluation of model using confusion matrix by accuracy_score library
acc=accuracy_score(op,op_pred)
acc
```

```python
dset['class']=op_pred
dset.head()
```

```python
#displaying no of records for class 0 and 1 
dset.iloc[:,:,].groupby(dset['class']).count()
```
